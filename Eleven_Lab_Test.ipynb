```python
import openai
import os
import elevenlabs
from IPython.display import Audio, display
import IPython.display as ipd

openai.api_key = os.getenv('OPENAI_API_KEY')
elevenlabs.set_api_key(os.getenv('ELEVENLABS_API_KEY'))

def generate_response(user_message):
    response = openai.ChatCompletion.create(
        model="gpt-4",  # Assurez-vous d'utiliser le bon identifiant de modle ici
        messages=[
            {"role": "system", "content": "Vous tes la copine de l'utilisateur, rpondez de manire concise et sexy. Ne dis pas que tu es une intelligence artificielle"},
            {"role": "user", "content": user_message},
        ],
        temperature=0.7,
        max_tokens=100
    )
    return response['choices'][0]['message']['content']

voice = elevenlabs.Voice(
        voice_id = 'WRLG2mTdj5mzd165xB5Y',
        settings = elevenlabs.VoiceSettings(
            stability = 0.8,
            similarity_boost = 0.7,
            style=0.9, 
            use_speaker_boost=True
        )
        )

for i in range(10):
    user_message = input("Vous: ")
    response = generate_response(user_message)
    audio = elevenlabs.generate(text=response, voice=voice, model="eleven_multilingual_v2")
    filename = f"margot_answer_{i+1}.mp3"
    elevenlabs.save(audio, filename)
    display(Audio(filename))
    
```
